```python
import pandas as pd
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Movies%20Recommendation.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df_features = df[['Movie_Genre', 'Movie_Keywords', 'Movie_Tagline', 'Movie_Cast', 'Movie_Director']].fillna('')
```

```python
df_features.shape
```

```python
df_features
```

```python
x = df_features['Movie_Genre'] + ' ' + df_features['Movie_Keywords'] + df_features['Movie_Tagline'] + ' ' + df_features['Movie_Cast'] + ' ' + df_features['Movie_Director']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
t = TfidfVectorizer()
```

```python
x = t.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ss = cosine_similarity(x)
```

```python
ss
```

```python
favm = input("Enter your favourite movie: ")
```

```python
allmovies = df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
movierec = difflib.get_close_matches(favm, allmovies)
```

```python
movierec
```

```python
closematch = movierec[0]
```

```python
closematch
```

```python
indexof_cm = df[df.Movie_Title == closematch]['Movie_ID'].values[0]
print(indexof_cm)
```

```python
rec_score = list(enumerate(ss[indexof_cm]))
rec_score
```

```python
ssm = sorted(rec_score, key = lambda x:x[1], reverse =True)
```

```python
print("Top 30 Movie Suggestions For You: \n")
i = 1

for movie in ssm:
  index = movie[0]
  title_from_index = df[df.index==index]['Movie_Title'].values[0]
  if (i<31):
    print(i, '-', title_from_index)
    i+=1
```
